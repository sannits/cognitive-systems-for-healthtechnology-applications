```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(1001)
x
```

```python
y = np.log10(x+1)
y
```

```python
plt.plot(x,y)
plt.title('Graph of y = log(x+1)')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.grid()
plt.show()
```

```python
#location of the dataset
filename = 'http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.switzerland.data'

#column names for the data
colnames = ["a", "b", "c", "d","e","f","g","h","i","j","k","l","m","n"]

#read the data to data frame
df = pd.read_csv(filename, 
                 names=colnames, 
                 na_values='?')
```

```python
df.head()
```

```python
# Plot the values of column H
df['h'].plot()
plt.show()
```

```python
# Calculate the descriptive statistics
df['h'].describe()
```
